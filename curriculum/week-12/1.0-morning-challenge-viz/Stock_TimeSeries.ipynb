```python
import pandas_datareader.data as web
from datetime import datetime as dt

start = dt(2000, 01, 01)
end = dt(2016, 12, 02)

spy = web.DataReader('SPY','yahoo', start, end)
spy.describe()
```

```python

spy.head()
```

```python
# day vs prior day, 2 days, 3 days

from datetime import timedelta
offset1 = timedelta(days=-1)
offset2 = timedelta(days=-2)
offset3 = timedelta(days=-3)

# spy['prior_day_open'] = spy['Open'].shift(-1)
spy['prior_day_high'] = spy['High'].shift(-1)
spy['prior_day_low'] = spy['Low'].shift(-1)
spy['prior_day_close'] = spy['Close'].shift(-1)

spy['prior_2day_open'] = spy['Open'].shift(-2)
spy['prior_2day_high'] = spy['High'].shift(-2)
spy['prior_2day_low'] = spy['Low'].shift(-2)
spy['prior_2day_close'] = spy['Close'].shift(-2)

spy['prior_3day_open'] = spy['Open'].shift(-3)
spy['prior_3day_high'] = spy['High'].shift(-3)
spy['prior_3day_low'] = spy['Low'].shift(-3)
spy['prior_3day_close'] = spy['Close'].shift(-3)

spy.head()
```

```python
spy['prior_2day_avg_close'] = sum((spy['Close'].shift(-1))+(spy['Close'].shift(-2)))
```
