## 3.  Perform a K-Means Clustering 

```python
%matplotlib inline 

import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from sklearn import cluster, datasets, preprocessing, metrics
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
```

#### 3.1 Split the dataset into the attribute data and class labels

```python
cols = df.columns
df_scaled = preprocessing.MinMaxScaler().fit_transform(df[cols])
```

```python
pd.DataFrame(df_scaled, columns=cols).describe()
```

#### 3.2 Conduct the k-means clustering

```python
k = 2
kmeans = cluster.KMeans(n_clusters=k)
kmeans.fit(df_scaled)
```

#### 3.3 Compute the labels and centroids

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
inertia = kmeans.inertia_
metrics.silhouette_score(df_scaled, labels, metric='euclidean')
```

```python
df['label'] = labels
df.head()
```
