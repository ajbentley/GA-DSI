```python
# From the new char DataFrame, select the height. Can you create a histogram?

char['height']
```

```python
# Create a histogram by creating a list comprehension + a function to turn the data into inches

def clean(x):
    if pd.isnull(x):
        return x
    elif 'Under' in x:
        return 78
    elif 'above' in x:
        return 60
    else:
        ft = int(x.split("'")[0]) * 12
        inches = int(x.split("'")[1].replace('"',''))
        return ft + inches
    


char['height'] = [clean(x) for x in char['height']]
```

```python
char['height'].hist();
```

```python
def count_degrees(x):
    degree_dict[x] = degree_dict[x] + 1
    
    
[count_degrees(x) for x in df['edu']]
```
